```python
import cinrad
%matplotlib inline
import numpy as np
import matplotlib.pyplot as mp
f = cinrad.io.CinradReader('Z_RADR_I_Z9576_20190810000600_O_DOR_SA_CAP.bin.bz2')
tilt_number = 0
data_radius = 230
data_dtype = 'REF' # stands for reflectivity
r = f.get_data(tilt_number, data_radius, data_dtype)

data = r.data
print(data.shape)
x1=[i for i in range(230)]
y1=[i for i in range(366)]
x,y=np.meshgrid(x1,y1)

z=data

# 
mp.figure('Contour', facecolor='lightgray')
mp.title('Contour', fontsize=14)
mp.xlabel('x', fontsize=12)
mp.ylabel('y', fontsize=12)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
# 
cntr = mp.contour(x, y, z, 8, colors='black',
                  linewidths=0.5)
mp.contourf(x, y, z, 8, cmap='jet')
mp.show()
```

```python
import cinrad
%matplotlib inline
import numpy as np
import matplotlib.pyplot as mp
from cinrad.visualize import PPI
f = cinrad.io.CinradReader('Z_RADR_I_Z9576_20190810000600_O_DOR_SA_CAP.bin.bz2')
tilt_number = 0
data_radius = 230
data_dtype = 'REF' # stands for reflectivity
r = f.get_data(tilt_number, data_radius, data_dtype)
v = []
v.append(r)
f =cinrad.easycalc.GridMapper(v)
grid = f(0.01)
print(grid.data)
print(type(f))
# PPI(grid)
data = grid.data
x1=[i for i in range(474)]
y1=[i for i in range(416)]
x,y=np.meshgrid(x1,y1)

z=data
print(z.shape)
print(grid.drange)
print(grid.lon)
print(grid.lat)
print(grid.scan_info)
print(grid.stp)
print(grid.scantime)
print(grid.dtype)
#
mp.figure('Contour', facecolor='lightgray')
mp.title('Contour', fontsize=14)
mp.xlabel('x', fontsize=12)
mp.ylabel('y', fontsize=12)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
# 
cntr = mp.contour(x, y, z, 8, colors='black',
                  linewidths=0.5)

mp.contourf(x, y, z, 8, cmap='jet')

mp.show()

```
